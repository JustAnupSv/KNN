```python
import pandas as pd
import numpy as np
from sklearn import preprocessing 
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Data Preprocessing

```python
label_encoder = preprocessing.LabelEncoder()
zoo["animal name"] = label_encoder.fit_transform(zoo["animal name"])
```

```python
zoo
```

```python
zoo_array = zoo.values
X = zoo_array[:, 1:17]
X
```

```python
Y = zoo_array[:, -1]
Y
```

```python
kfold = KFold(n_splits=4)
```

```python
model = KNeighborsClassifier(n_neighbors=13)
results = cross_val_score(model, X, Y, cv=kfold)
```

```python
print(results.mean())
```

## Algorithm Tuning

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
```

```python
print(grid.best_params_)
```

## CV Resluts

```python
import matplotlib.pyplot as plt 
get_ipython().run_line_magic('matplotlib', 'inline')
# Choosing k between 1 to 70
k_range = range(1, 70)
k_scores = []
# Using iteration to caclulate different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=4)
    k_scores.append(scores.mean())
# Plotting to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
# choose k between 1 to 40
k_range = np.array([2*i+1 for i in range(1,20)])
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=10)
    k_scores.append(scores.mean())
# plot to see clearly
plt.figure(figsize=(12,8))
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.xticks(k_range)
plt.show()
```
